```python
import pandas as pd
import numpy as np
```

```python
name=pd.Series(["小陈","小李","小王","小张","小胡","小唐"],index=["001","002","003","004","005","006"])
test1=pd.Series([85,73,82,91,65,94],index=["001","002","003","004","005","006"])
test2=pd.Series([77,67,92,81,68,89],index=["001","002","003","004","005","006"])
test3=pd.Series([84,87,72,61,94,91],index=["001","002","003","004","005","006"])
students=pd.DataFrame({"姓名":name,"考试1":test1,"考试2":test2,"考试3":test3})
students
```

```python
average=students.loc[:,"考试1":"考试3"].mean(axis=1)
name=students["姓名"]
students_average = pd.DataFrame({ "姓名":name , "平均分":average })
students_average
```

```python
students.loc[:,"考试1":"考试3"].apply( lambda x:np.sort(x)[-2] )
```

```python
def get_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "E"
```

```python
students.loc[:,"考试1":"考试3"].applymap(get_grade)
```
